```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
class RetailSalesAnalyzer:
  def clean_data(self,df):
    self.df=df.dropna()
    return self.df

  def analyze_data(self,df):
    df=self.clean_data(df)
    total_sales=df.groupby('Product')['Sales'].sum()
    best_selling = df['Product'].max()
    sales=np.array(df['Sales'])
    avg_sales=np.mean(sales)
    return {f'The total sale for Product A:{total_sales[0]}, Product B:{total_sales[1]}, Product C:{total_sales[2]}',
            f'The best selling product is {best_selling}',
            f'The average sale is {avg_sales}'}

  def sales_visualization(self,df):
    df['Date']=pd.to_datetime(df['Date'])
    date=df.groupby('Date')['Sales'].sum()
    date.plot(x='Date', y='Sales', kind='line')
    plt.xlabel('Date of the Sale')
    plt.ylabel('Sales')
    plt.show()

  def sales_per_product_visualization(self,df):
    try:
      sales_per_product=df.groupby('Product')['Sales'].sum()
      sales_per_product.plot(x='Product', y='Sales', kind='bar')
      plt.xlabel('Product')
      plt.ylabel('Sales')
      plt.show()
    except Exception as e:
      print(str(e))








```

```python
df=pd.read_csv('/content/retail_sales.csv')
#print(df)

retail_analyzer=RetailSalesAnalyzer()
res=retail_analyzer.analyze_data(df)
print(res)
visualization=retail_analyzer.sales_visualization(df)
sale_vis = retail_analyzer.sales_per_product_visualization(df)
```
